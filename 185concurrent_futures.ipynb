```python
from concurrent.futures import ThreadPoolExecutor, as_completed
from urllib.request import urlopen

urls = ['https://www.naver.com/',
        'https://www.cnn.com/',
        'https://www.daum.net/',
        'https://blog.naver.com/samsjang/',
        'https://www.python.org/']
```

```python
def loadUrl(url, timeout):
  with urlopen(url, timeout = timeout)as f:
    return f.read()
```

```python
with ThreadPoolExecutor(max_workers=5)as executor:
  fs = {executor.submit(loadUrl, url, 60):url for url in urls}
  for future in as_completed(fs):
    url = fs[future]
    try:
      data = future.result()
    except Exception as e:
      print(f'{url}예외 발생: {e}')
    else:
      print(f'{url}접속 성공: {len(data)}바이트 수신')
```

```python
def checkPrimeNumber(x):
  ret = False

  if x==2:
    return True
  if x<=1 or x%2==0:
    return False

  print(f'{x:,}가 소수인지 아닌지 판별 중...')
  stime = time()
  for i in range(3,int(x**(1/2))+1,2):
    if x%i ==0:
      ret = False
  else:
    ret = True
  etime = time()

  if ret:
    print(f'{x:,}는 소수입니다. 소수판별시간:{etime-stime:.2f}초')
  else:
    print(f'{x:,}는 소수가 아닙니다.')

  return ret
```
