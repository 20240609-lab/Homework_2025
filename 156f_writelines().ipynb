```python
data1 = 'This is data1'
data2 = 'This is data2'
with open('mydata.txt','w')as f:
  f.writelines(data1)
  f.writelines(data2)
```

```python
with open('mydata.txt','r')as f:
  data = f.read()
print(data)
```

```python
data1 = 'This is data1\n'
data2 = 'This is data2\n'
```

```python
data = ['This is data1\n','This is data2\n']
with open('mydata.txt','w')as f:
  f.writelines(data)
```

```python
with open('mydata.txt','r')as f:
  data = f.read()
print(data)
```
