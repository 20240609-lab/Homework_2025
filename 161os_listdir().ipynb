```python
from os import listdir

folder = './'
ret = listdir(folder)
print(ret)
```

```python
form glob import glob

target = '*.txt'
ret = glob(target)
print(ret)
```

```python
target = '**'
ret = glob(target, recursive=True)
ret.sort()
for r in ret:
  print(r)
```
