```python
class Character:
  def __init__(self):
    self.strength = 1
    self.offense_power = 1
    self.intelligence = 1
    self.mentality = 1
    self.agility = 1
```

```python
class Wizard(Character):
  def setBasic(self, name = None, level=1,hp=100):
    self.name = name
    self.level = level
    self.hp = hp
```

```python
player1 = Wizard()
player1.setBasic('StormBaby')
print(f'[{player1.name}]레벨:{player1.level}')
print(f'정신력:{player1.mentality}')
```

```python
class Wizard(Character):
  def __init__(self, name = None):
    super().__init__()
    self.name = name
    self.level = 1
    self.hp = 100
```

```python
player1 = Wizard('StormBaby')
print(f'[{player1.name}]레벨:{player1.level}')
print(f'정신력:{player1.mentality}')
```
