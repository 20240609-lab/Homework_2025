```python
class Wizard:
  generated = 0

  def __init__(self,name):
    Wizard.generated += 1
    self.name = name
    self.level = 1
    self.hp = 100
    self.initAbility()

  def initAbility(self):
    self.strength = 1
    self.offense_power = 1
    self.intelligence = 10
    self.mentality = 10
    self.agility = 1
```

```python
player1 = Wizard('FirePlayer')
print(f'마법사 생성 수 : {player1.generated}\n캐릭터 이름: {player1.name}')
```

```python
player2 = Wizard('GodGo')
print(f'마법사 생성 수 : {player2.generated}\n캐릭터 이름: {player2.name}')
```
