```python
data = input('저장할 문장을 입력하세요: ')
flag = input('입력한 문장을 저장하시겠습니까? (y/n)')
if flag.lower() == 'y':
  with open('mydata.txt','w')as f:
    f.write(data)
  print('입력한 문장을 "mydata.txt"에 저장했습니다.')
else:
  print('입력한 문장을 저장하지 않습니다.')
```

```python
data = b'A lot of things occur each day'
with open('mydata.bin','wb')as f:
  f.write(data)
```
