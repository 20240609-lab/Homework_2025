```python
def splitFilename(file):
  buf = file.split('.')
  if len(buf) > 1:
    ext = buf[-1]
    fname = '.'.join(buf[:-1])
  else:
    fname = buf[0]
    ext = None
  return (fname, ext)
```

```python
##import splitfilename

filename = 'python.200.study.py'
ret = splitFilename(filename)
print(f'파일 이름: {ret[0]},파일확장자: {ret[1]}')
```
