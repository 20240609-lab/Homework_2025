```python
def cumalativeSum(n):
  ret = 0
  for i in range(1,n+1):
    ret += i
  return ret
```

```python
from time import time

n = 10000000
stime = time()
answer = cumalativeSum(n)
etime = time()
print(f'1에서 {n:,}까지 누적합 계산 시간: {(etime-stime):.3f}초')
```
