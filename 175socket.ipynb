```python
import socket

HOST = ''
PORT = 7777

def echoServer():
  with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sock:
    sock.bind((HOST, PORT))
    sock.listen()
    print('클라이언트 연결을 기다린는 중...')
    conn, addr = sock.accept()
    with conn:
      print(f'{addr[0]}와 연결됨')
      while True:
        data = conn.recv(1024)
        if not data:
          break
          print(f'수신 메시지: {data.decode()}')
          conn.sendall(data)

echoServer()
```
