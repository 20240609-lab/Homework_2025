```python
names = {'Mary':10999,'Cris':2111,'Aimy':9778}
population = names.setdefault('Sams')
print(f'names:{names}\npopulation:{population}')
```

```python
population = names.setdefault('Sams',100)
print(f'names:{names}\npopulation:{population}')
```
