```python
intList = range(10)
idx = 10
try:
  print(intList[idx])
except Exception as e:
  exceptionClass = e.__class__.__name__
  exceptionDetail = e.__str__()
  print(f'예외 종류: {exceptionClass}, 예외내용:{exceptionDetail}')
```
