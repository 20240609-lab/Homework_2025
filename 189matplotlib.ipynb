```python
import matplotlib.pyplot as plt
import math
```

```python
krfont = {'family':'NanumBarunGothic', 'weight':'bold', 'size':10}
matplotlib.rc('font', **krfont)
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
PI = 3.141592
xs, ys1, ys2 = [],[],[]
for x in range(360):
  xs.append(x)
  ys1.append(math.sin(PI*x/180))
  ys2.append(math.cos(PI*x/180))
```

```python
plt.plot(xs,[0]*len(xs), color = 'black')
plt.plot(xs, ys1, label = '사인 곡선')
plt.plot(xs, ys2, label = '코사인 곡선')
plt.xlim([0,360])
plt.xlabel('X-값')
plt.ylabel('Y-값')
plt.title('삼각함수')
plt.legend()
plt.show()
```
