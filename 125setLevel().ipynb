```python
def setLevel(character,level,*args):
  print(f'장수:{character},레벨:{level}')
  if len(args) != 0:
    print(f'기타 특성:{args}')
  return
```

```python
setLevel('조조',3)
```

```python
setLevel('유비',3,'named','leader')
```

```python
def setLevel(character,level,**kwargs):
  print(f'장수:{character},레벨:{level}')
  if len(kwargs) != 0:
    for k, v in kwargs.items():
      print(f'{k}:{v}')
  return
```

```python
setLevel('조조',3)
```

```python
setLevel('유비',3,category='named',position='leader')
```
