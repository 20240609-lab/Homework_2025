```python
import socketserver

HOST = ''
PORT = 7777

class MyTcpHandler(socketserver.BaseRequestHandler):
  def handle(self):
    print(f'[{self.client_address[0]}] 연결됨')

    try:
      while True:
        self.data = self.request.recv(1024)
        if self.data.decode()=='/q':
          print(f'[{self.client_address[0]}] 사용자 종료')
          return
          print(f'수신 메시지 [{slef.data.decode()}]를 에코 전송')
          slef.request.sendall(self.data)
    except Exception as e:
      print(e)

def runServer():
  print('+++ 에코 서버를 시작합니다.')
  print('+++ 에코 서버를 끝내려면 Ctrl-C를 누르세요.')

  try:
    server = socketserver.TCPSever((HOST, PORT), MyTcpHandler)
    server.sever_forever()
  except KeyboardInterrupt:
    print('---에코 서버를 종료합니다.')

runServer()
```
