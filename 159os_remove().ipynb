```python
from os import remove


target = 'mydata.txt'
k = input(f'<{target}>을 삭제하시겠습니까? (y/n)')
if k.lower() == 'y':
  remove(target)
  print(f'<{target}>을 삭제했습니다.')
else:
  print(f'<{target}>을 삭제하지 않았습니다.')
```
